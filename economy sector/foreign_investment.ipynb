```python
PARTIES_TENTURE=[(0,5,'B'),(4,14,'C'),(14,22,'B')]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
investment_df = pd.read_csv("../FII INFLOW INTO INDIAN MARKETS.csv")
investment_df
```

```python
import matplotlib.pyplot as plt
YEARS = investment_df["Financial Year"]
VALUES=  investment_df["Total Investments (In INR crores) into equity,debt,hybrid"]
COLORS=[]
for year in YEARS:
    for tenure in PARTIES_TENTURE:
        year=int(year)
        if(year>=YEARS[tenure[0]] and year<=YEARS[tenure[1]-1]):
            if(tenure[2]=='B'):
                COLORS.append("red")
            else:
                COLORS.append("blue")
plt.bar(YEARS, VALUES,color=COLORS)
plt.title("USD to INR YEAR WISE DATA")
plt.xlabel("Timeline")
plt.ylabel("USD/INR")
colors = {'BJP':'red', 'CONGRESS':'blue'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels)
    
```
