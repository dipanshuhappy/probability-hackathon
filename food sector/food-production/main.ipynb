```python
#intialize Constants
PARTIES_TENTURE=[(0,4,'B'),(3,12,'C'),(11,14,'B')]
```

```python
import pandas as pd

food_average_yeild = pd.read_csv("../../All India level Average Yield of Principal Crops from 2001-02 to 2015-16.csv")
food_average_yeild
```

#Getting The average of the food product using panda in below code


```python
AVERAGE_FOOD_YIELD=food_average_yeild.sum(axis=1)
AVERAGE_FOOD_YIELD
```

```python
#Normalizaing Years
YEARS=[ x[:4] for x in food_average_yeild['Year']]
```

## Below is visualization of food production from 2001 to 2016 
Note - Red represents BJP and Blue represent Congress

```python
import matplotlib.pyplot as plt
#print(years)
legend_one_shown=False
leegend_two_shown=False
for tenure in PARTIES_TENTURE:
 if(tenure[2]=='B'):
   plt.plot(YEARS[tenure[0]:tenure[1]]
   ,AVERAGE_FOOD_YIELD[tenure[0]:tenure[1]],
   color='red',label='BJP')
   if(legend_one_shown==False):
      plt.legend()
      legend_one_shown=True
   

 elif (tenure[2]=='C'):
   plt.plot(YEARS[tenure[0]:tenure[1]],
   AVERAGE_FOOD_YIELD[tenure[0]:tenure[1]],
   color='blue',label='Congress')
   plt.legend()
plt.xticks(rotation=90)
plt.show()
```
