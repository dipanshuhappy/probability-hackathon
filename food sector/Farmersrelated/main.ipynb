```python
#intialize Constants
PARTIES_TENTURE=[(0,5,'B'),(4,15,'C'),(14,21,'B')]
```

```python
import pandas as pd

farmers_related = pd.read_csv("../../Farmer Suicide from 2000 to 2020.csv")
farmers_related
```

```python
farmers_related["Year"]
farmers_related["Farmer Suicide"]
```

```python
import matplotlib.pyplot as plt
#print(years)
YEARS=farmers_related["Year"]
legend_one_shown=False
leegend_two_shown=False
for tenure in PARTIES_TENTURE:
 if(tenure[2]=='B'):
   plt.plot(YEARS[tenure[0]:tenure[1]]
   ,farmers_related["Farmer Suicide"][tenure[0]:tenure[1]],
   color='red',label='BJP')
   if(legend_one_shown==False):
      plt.legend()
      legend_one_shown=True
   

 elif (tenure[2]=='C'):
   plt.plot(YEARS[tenure[0]:tenure[1]],
   farmers_related["Farmer Suicide"][tenure[0]:tenure[1]],
   color='blue',label='Congress')
   plt.legend()
plt.xticks(rotation=90)
plt.show()
```
